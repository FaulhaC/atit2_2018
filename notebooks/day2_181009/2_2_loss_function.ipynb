```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error as mse

plt.rcParams['figure.figsize'] = [15, 8]
```

```python
boston = tf.keras.datasets.boston_housing.load_data()[0]
X = boston[0]
Y = boston[1]
```

```python
c = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "black", "lstat", 'medv']
boston = pd.DataFrame({c: X[:, i] for i, c in enumerate(c[:-1])})
boston[c[-1]] = Y
```

```python
def get_hypothesis(theta_1, theta_2):
    return lambda x: theta_1 + x * theta_2
```

```python
x = boston.rm
y = boston.medv

# calculates mean squared error for a given choice of theta
def get_loss(theta_0, theta_1):
    h = get_hypothesis(theta_0, theta_1)
    y_pred = x.map(h)
    return mse(y_true=y, y_pred=y_pred)
```

```python
theta_1 = 20
theta_0 = -45

h_1 = get_hypothesis(theta_0, theta_1)

y_pred_1 = x.map(h_1)
```

```python
mse_1 = mse(y_true=y, y_pred=y_pred_1)
mse_1
```

```python
theta_1 = 15
theta_0 = -45
mse_2 = get_loss(theta_0, theta_1)
mse_2
```

```python
theta_1 = 10
theta_0 = -45
mse_3 = get_loss(theta_0, theta_1)
mse_3
```

```python
theta_1 = 11
theta_0 = -45

mse_4 = get_loss(theta_0, theta_1)
mse_4
```

```python
plt.rcParams['figure.figsize'] = [15, 8]

plt.scatter([20, 15, 10, 9], [mse_1, mse_2, mse_3, mse_4])
plt.ylabel('mse of hypothesis')
plt.xlabel('theta_2')
```

```python
theta_1s = list(range(-25, 40, 1))
```

```python
theta_0 = -30

theta_1s = list(range(-25, 40, 1))

losses = [get_loss(theta_0, theta_1) for theta_1 in theta_1s]
plt.scatter(theta_1s, losses)
plt.xlabel('theta_2')
plt.ylabel('mean squared error')
```
