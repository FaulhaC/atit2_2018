```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

params = {'legend.fontsize': 'x-large',
          'figure.figsize': (15, 10),
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'x-large',
         'ytick.labelsize':'x-large'}
plt.rcParams.update(params)
```

```python
boston = tf.keras.datasets.boston_housing.load_data()[0]
X = boston[0]
Y = boston[1]
```

```python
c = ["crim", "zn", "indus", "chas", "nox", "rm", "age", "dis", "rad", "tax", "ptratio", "black", "lstat", 'medv']
boston = pd.DataFrame({c: X[:, i] for i, c in enumerate(c[:-1])})
boston[c[-1]] = Y
```

```python
def get_hypothesis(theta_0, theta_1):
    def hypothese(x):
        return theta_0 + x * theta_1
    return hypothese

# or using lambda functions (fancy stuff!)
def get_hypothesis(theta_0, theta_1):
    return lambda x: theta_0 + x * theta_1
```

```python
h1 = get_hypothesis(-10, 3)
h2 = get_hypothesis(-10, 5)
h3 = get_hypothesis(-10, 10)
```

```python
h1(5), h2(5), h3(5)
```

```python
def plot(hs, colors):
    """
    plots a scatter plot of number of rooms (x-axis) and price (y-axis)
    and adds a list auf linear hypotheses (hs) in different colors
    """
    plt.figure()
    x = plt.scatter(boston.rm, boston.medv, c='black')
    plt.scatter(boston.rm, boston.medv, c='black')
    for i, h in enumerate(hs):
        plt.plot([3, 9], [h(3), h(9)], 'k-', color = colors[i])
    plt.ylim(0, 60)
    plt.xlabel('number of rooms')
    plt.ylabel('median value in $1000s')
    plt.show()
```

```python
# test get_hypothesis

theta_1 = 20
theta_0 = -45

h = get_hypothesis(theta_0, theta_1)

h(0), h(2), h(4), h(8)
```

```python
theta_1 = 20
theta_0 = -45

h_1 = get_hypothesis(theta_0, theta_1)

plt.rcParams['figure.figsize'] = [15, 8]

plot([h_1], ['red'])
```

```python
theta_1 = 15
theta_0 = -45

h_2 = get_hypothesis(theta_0, theta_1)


plot([h_1, h_2], ['red', 'purple'])
```

```python
theta_1 = 10
theta_0 = -45

h_3 = get_hypothesis(theta_0, theta_1)


plot([h_1, h_2, h_3], ['red', 'purple', 'blue'])
```
